```python
import pandas as pd
```

```python
df = pd.read_csv('../Data/Full_Data.csv')
df = df.drop(columns=['Unnamed: 0'])
df.rename(columns={'Prop_a':"Prop_a_1",
                   "Prop_b":"Prop_b_1",
                   "Prop_x":"Prop_x_1"}, inplace=True)
```

```python
# Define the desired column order
new_order = ['Name', 'R_a', 'Prop_a_1', 'Prop_a_2', 'R_b', 'Prop_b_1', 'Prop_b_2', 'Prop_b_3','R_x', 
             'Prop_x_1', 'Prop_x_2', 'Tolerance_Factor', 'Structure_of_Material', 'BandGap']

# Reorder the columns
df = df[new_order]
```

```python
df[df["Prop_b_2"] != 0]
```

```python
df['Prop_a_2'] = 1 - df['Prop_a_1']
df['Prop_b_3'] = 1 - df['Prop_b_1'] - df['Prop_b_2']
df['Prop_x_2'] = 1 - df['Prop_x_1']
```

```python
df.to_csv('../Data/Full_Data_ranamed_columns.csv', index=False)
```
